```python
import tkinter
import tkinter.messagebox
```

```python
def main():
    flag = True
    
    # change the tag
    def change_label_text():
        nonlocal flag
        flag = not flag
        color, msg = ('red', 'Hello, world!') if flag else ('blue','Goodbye, world')
        label.config(text= msg, fg = color)
        
    def confirm_to_quit():
        if tkinter.messagebox.askokcancel("Are you sure you want to quit?"):
            top.quit()
    
    # create top window
    top = tkinter.Tk()
    # size of window
    top.geometry('240x160')
    # title of window
    top.title("small game")
    # create label and add to top window
    label = tkinter.Label(top, text='Hello, world', font='Arial -32', fg='red')
    label.pack(expand=1)
    # create a button
    panel = tkinter.Frame(top)
    # create buttion object, assign to a containter, band return function through 'command'
    button1 = tkinter.Button(panel, text="change", command=change_label_text)
    button1.pack(side='left')
    
    button2 =tkinter.Button(panel, text='exit', command=confirm_to_quit)
    button2.pack(side='right')
    panel.pack(side='bottom')
    tkinter.mainloop()
    
if __name__ == '__main__':
    main()
```

```python
import pygame

def main():
    # initiate
    pygame.init()
    # initialize window
    screen = pygame.display.set_mode((800, 600))
    # set window title
    pygame.display.set_caption("Big ball and samll ball")
    # set window color
    screen.fill((242, 242, 242))
    # draw a circle(screen, color, center, radius, filled)
    pygame.draw.circle(screen, (255, 0, 0), (100, 100), 30 , 0)
    # refresh the current window
    pygame.display.flip()
    running = True
    
    # circulation
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
if __name__ =="__main__":
    main()
```

```python
"Animation"
import pygame

def main():
    pygame.init()
    screen = pygame.display.set_mode((600, 800))
    pygame.display.set_caption("Big")
    x,y = 50, 50
    running = True
    while running:
        for event in pygame.event.get():
            if event.type ==pygame.QUIT:
                running = False
        screen.fill((255, 255, 255))
        pygame.draw.circle(screen, (255, 0, 0), (x,y), 30, 0)
        pygame.display.flip()
        pygame.time.delay(50)
        x, y = x + 5, y
main()
```

```python
from enum import Enum, unique
from math import sqrt
from random import randint

import pygame

@unique
class Color(Enum):
    RED = (255, 0, 0)
    GREEN = (0, 255, 0)
    BLUE = (0, 0, 255)
    BLACK = (0, 0, 0)
    WHITE = (255, 255, 255)
    GRAY = (242, 242, 242)
    
    @staticmethod
    def random_color():
        r = randint(0, 255)
        g = randint(0, 255)
        b = randint(0, 255)
        return (r,g,b)
    
class Ball(object):
    def __init__(self, x, y, radius, sx, sy, color=Color.RED):
        self.x = x
        self.y = y
        self.radius = radius
        self.sx = sx
        self.sy = sy
        self.color = color
        self.alive = True
        
    def move(self, screen):
        self.x += self.sx
        self.y += self.sy
        if self.x - self.radius <=0 or \
            self.x +self.radius >= screen.get_width():
            self.sx = -self.sx
        if self.y - self.radius <=0 or \
            self.y + self.radius >=screen.get_height():
            self.sy = -self.sy
    
    def eat(self, other):
        if self.alive and other.alive and self!= other:
            # distance between two balls
            dx, dy = self.x - other.x, self.y -other.y
            distance = sqrt(dx ** 2 + dy ** 2)
            
            if distance < self.radius+other.radius and self.radius>other.radius:
                other.alive = False
                self.radius = self.radius + int(other.radius*0.146)
                
    def draw(self, screen):
        pygame.draw.circle(screen, self.color, (self.x,self.y), self.radius, 0)
    
    
    
def main():
    balls = []
    pygame.init()
    screen = pygame.display.set_mode((800, 600))
    pygame.display.set_caption("Big ball eats Small ball")
    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
        
            if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:
                x, y = event.pos
                radius = randint(10, 100)
                sx, sy = randint(-10, 10), randint(-10, 10)
                color = Color.random_color()
                ball = Ball(x,y,radius, sx, sy, color)
                #相同名字的不同实例可以相加？
                # 不同object存在不同的地址中
                balls.append(ball) 
        screen.fill((255, 255, 255))
        
        for ball in balls:
            if ball.alive:
                ball.draw(screen)
            else:
                balls.remove(ball)
        pygame.display.flip()
        
        pygame.time.delay(50)
        for ball in balls:
            ball.move(screen)
            for other in balls:
                ball.eat(other)
                

if __name__ == '__main__':
    main()  
```

```python
class Student(object):
    def __init__(self, name, age):
        self.name = name
        self.age = age

stus =[]
names=["kevin", "jimmy", "Lucy"]
ages = [10, 15, 20]
for name in names:
    for age in ages:
        stu = Student(name, age)
        stus.append(stu)
for i in range(len(stus)):
    print(stus[i], end="\n")
```
